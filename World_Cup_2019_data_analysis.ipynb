## 4. Exploratory Data Analysis (EDA)

### 4.1 Total Matches Played in the World Cup

Q. Total number of matches played in the dataset.

```python
plt.figure( figsize = (3, 6))
sns.barplot(x = ["2019 World Cup"], y = df.shape[0])
plt.title(f"Total Matches Played In 2019 World Cup \n")
plt.show()
```

This bar plot shows the total number of matches played during the ICC Cricket World Cup 2019.
As we can see, a total of 48 matches were played in the tournament, including the group stage, semi-finals, and the final.

### 4.2 Matches Ended Without A Result

Q. How many matches ended without a result

```python
no_result = 0
for result in df["Result"]:
    if result in ["No result", "Match abandoned"]:
        no_result += 1
print("Total matches ended without a result:", no_result)
```

Out of the 48 matches played in the 2019 Cricket World Cup, 4 matches were either not completed or abandoned.
This indicates that a small portion of matches were affected by interruptions, such as weather conditions, which is important when analyzing tournament outcomes and team performance.

### 4.3 Stadium Housting Most Matches

Q. Which stadium hosted the most matches?

```python
plt.figure( figsize = (6, 6))
sns.countplot(y = "Ground / Location", data = df, order = df["Ground / Location"].value_counts().index)
plt.title("Stadium hosting most matches")
plt.xlabel("Matches Played")
plt.show()
```

This bar plot shows the number of matches hosted by each stadium during the 2019 Cricket World Cup.
It helps identify the most frequently used venues throughout the tournament.
From the visualization, we can observe that some grounds, such as Old Trafford, Manchester hosted more matches than others, Indicating their importance as key venues for the event.
